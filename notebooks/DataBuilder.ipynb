# Importando os dados do DataSus

```python
df_cap = pd.read_csv('../BASE_DE_DADOS_CNES_202003/dataset/datasus.csv', encoding='latin1')
grouped = df_cap[['NO_BAIRRO', 'QT_EXIST', 'QT_SUS']].groupby(['NO_BAIRRO']).sum()
```

```python
grouped
```

Concatenando

```python
join_df = df_usu.set_index('bairro_resid__estadia').join(grouped, how='inner').reset_index()
join_df.rename(columns={'index': 'bairro'}, inplace=True)
join_df = pd.concat([join_df, pd.get_dummies(join_df['evolução'])], axis=1)
join_df['casos_novos'] = join_df[['ativo', 'recuperado', 'óbito']].sum(axis=1)
join_df['casos_totais'] = 0
join_df = pd.concat([join_df, pd.get_dummies(join_df['faixa_etária'], prefix='idade')], axis=1)
join_df = pd.concat([join_df, pd.get_dummies(join_df['sexo'], prefix='sexo')], axis=1)
join_df = join_df[join_df['sexo_I']==0].drop(columns=['sexo_I'])
join_df
```

```python
join_df.columns
```

```python
x1 = join_df.groupby(['bairro', 'dt_notific'])[['casos_novos', 'casos_totais', 'ativo', 'recuperado', 'óbito', 'sexo_F', 'sexo_M', 'idade_De 0 a 9', 'idade_De 10 a 19',
       'idade_De 100 a 109', 'idade_De 110 a 119', 'idade_De 20 a 29',
       'idade_De 30 a 39', 'idade_De 40 a 49', 'idade_De 50 a 59',
       'idade_De 60 a 69', 'idade_De 70 a 79', 'idade_De 80 a 89',
       'idade_De 90 a 100', 'idade_N/D']].sum()
```

```python
x2 = join_df.groupby(['bairro', 'dt_notific'])[['QT_EXIST', 'QT_SUS']].mean()
x2 = x2.rename(columns={'QT_EXIST': 'n_leitos', 'QT_SUS': 'n_leitos_sus'})
```

```python
x = pd.concat([x2, x1], axis=1)
x['casos_totais'] = x['casos_novos'].groupby(level=0).cumsum()
x
```

```python
x.to_csv('../BASE_DE_DADOS_CNES_202003/dataset/data.csv')
```
