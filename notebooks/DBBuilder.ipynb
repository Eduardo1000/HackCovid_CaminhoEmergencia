```python
import pandas as pd
import os
pd.set_option('max_columns', 100)
```

```python
df_usu = pd.read_csv('../BASE_DE_DADOS_CNES_202003/dataset/sivep_gripe_2020.csv', 
                     encoding='latin1', sep=';')
print(df_usu.shape)
df_usu.head()
```

## Removendo colunas que tenham mais 80% das variáveis nulas

```python
aux_df = df_usu.isnull().sum()
columns_with_many_nulls = aux_df[aux_df>0.8*df_usu.shape[0]].keys().tolist()
df_usu = df_usu[df_usu.columns[~df_usu.columns.isin(columns_with_many_nulls)]]
print(df_usu.shape)
df_usu.head(3)
```

## Seleciona apenas Pacientes que foram internados no municipio do Rio de Janeiro

```python
df_usu = df_usu[df_usu['Código do município de internação']==330455]
print(df_usu.shape)
df_usu.head(3)
```

## Removendo colunas que tenham pouca variabilidade

```python
columns_without_variance = []
for column in df_usu.columns:
    if df_usu[column].nunique() == 1:
        columns_without_variance.append(column)
df_usu.drop(columns=columns_without_variance, inplace=True)
df_usu
```
